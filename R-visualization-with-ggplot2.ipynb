## Creating a simple ggplot
As the first example let's look at the `mpg` dataset from *ggplot2* package and try to answer a question about the relationship between a car's engine size and its fuel comsumption.

### The mpg data.frame
`mpg` (aka `ggplot2::mpg`) contains fuel economy data from 1999 and 2008 for 38 popular models of car. To find out more about this data frame one can use `?mpg` command. A data frame is a rectangular collection of variables (in the columns) and observations (in the rows).

```r
head(mpg) # looking at the first rows of the data frame
```

```r
# Before we start plotting, let's change the default size of our plots
options(repr.plot.width=6, repr.plot.height=4.5)
```

Question: What does the relationship between engine size and fuel efficiency look like?

Variables of interest are:

`displ`: engine displacement, in litres

`hwy`: highway miles per gallon

To plot `mpg`, run this code to put `displ` on the x-axis and `hwy` on the y-axis:

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy))
```

The plot shows that for the most part cars with bigger engine consume more fuel.

We begin a plot by the function `ggplot()`. It creates a coordinate system that you can add layers to. The first argument is the dataset that we want to use for plotting. `ggplot(data = mpg)` creates an empty graph. We complete the graph by adding one or more layers to it. In this example we are adding one layer by `goem_function()` to graph a scatterplot. *ggplot2* comes with many `geom` (geometrical) functions and each add a different type of layer.

Each `geom` in *ggplot2* takes a `mapping` argument. This defines how variables in our dataset are mapped to visual properties. The `mapping` argument is always paired with `aes()`, and the `x` and `y` arguments of `aes()` specify which variables to map to the `x` and `y` axes. *ggplot2* looks for the mapped variable in the `data` argument, in this case, `mpg`.

**A graphing template**

To make a graph, replace the bracketed sections in the code below with a dataset, a geom function, or a collection of mappings.

`ggplot(data = <DATA>) + 
  <GEOM_FUNCTION>(mapping = aes(<MAPPINGS>))`

## Aesthetic mappings
Looking at the graph we just made, we see that there are a few models that despite their engine size are relatively efficient (~25 mpg) and seem to fall out of the linear trend (image below). How can we exmplain these cars? 

<img src="https://github.com/soltaniehha/online_content/blob/master/png/mpg_outliers.png?raw=true" width="578px" style="float: left; margin-left: 5px;">

We might be able to exmpain it by their class attribute, after all not all large engine vehicles are SUV. Let's start by using `class` column to color the points. To do this we need to map this field to an **aesthetic**. An aesthetic is a visual property of the objects in our plot. Aesthetics include things like the size, the shape, or the color of our points.

```r
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy, color = class))
```

The colors reveal that many of the unusual points are two-seater cars. These sports cars have large engines like SUVs and pickup trucks, but small bodies like midsize and compact cars, which improves their gas mileage.

#  Multilayer scatterplot

To illustrate an example we will use another ggplot2 dataset, called *diamonds*. We will plot a relationship between weight (in carat) and price of 1000 diamonds. For the purpose of this illustration a small dataset will suffiece, so we will use a subset:

```r
# row count of the original dataset
nrow(diamonds)
```

```r
set.seed(100) # setting seed will ensure the reproducibility of the outcome
subset_diamonds = diamonds[sample(nrow(diamonds), 1000), ] # producing a random subset with 1000 samples
head(subset_diamonds)
```

```r
ggplot() + 
layer(
    data = subset_diamonds, mapping = aes(x = carat, y = price),
    geom = "point", stat = "identity", position = "identity"
) +
scale_y_continuous() +
scale_x_continuous() +
coord_cartesian()
```

*ggplot2* assumes many defaults in a smart way. Intelligent defaults allow us to simplify the specification. We could reduce the code to a much simpler one, similar to what we saw in the earlier graphs, and still get an identical outcome. For instance: 
1. We only need specify one of geom and stat, as each geom has a default stat (and vice versa).
2. The Cartesian coordinate system is used by default, so does not need to be specified.
3. Default scales will be added according to the aesthetic and type of variable.
4. Typically, we will specify a default dataset and mapping in the ggplot call, use position-based matching in aes, and use a shorthand for the layer

```r
ggplot(subset_diamonds, aes(carat, price)) +
geom_point()
```

It appears that price and weight have a logarithmic relationship, to confirm we can change the default linear-transformatin to a log-transformation. We also add a second layer to draw this hypothetical linear relationship between the log-transformed variables. This is done by `stat_smooth()`

```r
ggplot(subset_diamonds, aes(carat, price)) +
geom_point() +
stat_smooth(method = lm) +
scale_x_log10() +
scale_y_log10()
```

which is shorthand for:

`ggplot() +
layer(
    data = diamonds, mapping = aes(x = carat, y = price),
    geom = "point", stat = "identity", position = "identity"
) +
layer(
    data = diamonds, mapping = aes(x = carat, y = price),
    geom = "smooth", position = "identity",
    stat = "smooth", method = lm
) +
scale_y_log10() +
scale_x_log10() +
coord_cartesian()`
